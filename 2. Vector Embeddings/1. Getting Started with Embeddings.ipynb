# HuggingFace Embeddings

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
text = "This is a sample text to test sentence embeddings."
sentence_embedding = embeddings.embed_query(text=text)

print(f"Embedding length: {len(sentence_embedding)}")
```

```python
text = "This is a sample text to test sentence embeddings. It contains two sentences"
list_texts = text.split(".")

sentence_embedding = embeddings.embed_documents(texts=list_texts)

print(f"Embedding length: {len(sentence_embedding)}")
print(f"Embedding length: {len(sentence_embedding[0])}")
```

# OpenAI Embeddings

```python
from langchain_openai import OpenAIEmbeddings
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
```

```python
text = "This is a sample text to test OpenAI embeddings."
sentence_embedding = embeddings.embed_query(text=text)

print(f"Embedding length: {len(sentence_embedding)}")
```
